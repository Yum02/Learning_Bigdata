4. 스타일

```python
import matplotlib.pyplot as plt
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y)
```

## 마커

```python
plt.plot(x, y, linewidth=5)
```

```python
plt.plot(x, y, marker='o')
```

```python
plt.plot(x, y, marker='o', linestyle='None')
```

```python
plt.plot(x, y, marker='v')
```

```python
plt.plot(x, y, marker='v', markersize=10)
```

```python
plt.plot(x, y, marker='X', markersize=10)
```

```python
plt.plot(x, y, marker='v', markersize=10, markeredgecolor='red')
```

```python
plt.plot(x, y, marker='o', markersize=20, markeredgecolor='red')
```

```python
plt.plot(x, y, marker='o', markersize=20, markeredgecolor='red', markerfacecolor='yellow')
```

## 선 스타일

```python
plt.plot(x, y, linestyle=':')
```

```python
plt.plot(x, y, linestyle='--')
```

```python
plt.plot(x, y, linestyle='-.')
```

```python
plt.plot(x, y, linestyle='-')
```

## 색깔

```python
plt.plot(x, y, color='pink')
```

```python
plt.plot(x, y, color='#ff0000')
```

```python
plt.plot(x, y, color='b')
```

## 포맷

```python
plt.plot(x, y, 'ro--') ##color, marker, linestyle
```

```python
plt.plot(x, y, 'bv:')
```

```python
plt.plot(x, y, marker='o', mfc='red', ms=10, mec='blue', ls=':')
```

```python
plt.plot(x, y, marker='o', mfc='red', ms=10, alpha=0.7) #slpha : 투명도(0~1)
```

## 그래프 크기

```python
plt.figure(figsize=(10, 5))
plt.plot(x, y)
```

```python
plt.figure(figsize=(5, 10))
plt.plot(x, y)
```

```python
plt.figure(figsize=(10, 5), dpi=50) #그래프 확대 조절
plt.plot(x, y)
```

```python
plt.figure(facecolor='yellow')
plt.plot(x, y)
```
