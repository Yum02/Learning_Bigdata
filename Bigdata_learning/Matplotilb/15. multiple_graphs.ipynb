15. 여러 그래프

```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_excel('../Pandas/score1.xlsx')
df
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 10)) # 2*2 plot
fig.suptitle('Multiple Graph')

#[0, 0]의 그래프
axs[0, 0].bar(df['Name'], df['KR'], label='KR Score') #데이터 설정
axs[0, 0].set_title('First Graph') #제목
axs[0, 0].legend() #범례
axs[0, 0].set(xlabel='Name', ylabel='Score')
axs[0, 0].set_facecolor('lightyellow') #전경색
axs[0, 0].grid(linestyle='--', linewidth=0.5)

#[0, 1]의 그래프
axs[0, 1].plot(df['Name'], df['Math'], label='Math')
axs[0, 1].plot(df['Name'], df['EN'], label='EN')
axs[0, 1].legend()

#[1, 0]의 그래프
axs[1, 0].barh(df['Name'], df['Height'])

#[1, 1]의 그래프
axs[1, 1].plot(df['Name'], df['SC'], alpha = 0.5)
```
