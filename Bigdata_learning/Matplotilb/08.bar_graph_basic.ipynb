8. 막대그래프 기본

```python
import matplotlib.pyplot as plt
```

```python
labels = ['A', 'B', 'C']
values = [190, 187, 184]

plt.bar(labels, values)
```

```python
labels = ['A', 'B', 'C']
values = [190, 187, 184]

plt.bar(labels, values, color='r')
```

```python
labels = ['A', 'B', 'C']
values = [190, 187, 184]
colors = ['r', 'g', 'b']

plt.bar(labels, values, color=colors, alpha=0.5)
```

```python
labels = ['A', 'B', 'C']
values = [190, 187, 184]

plt.bar(labels, values)
plt.ylim(175, 195)
```

```python
plt.bar(labels, values, width=0.5)
```

```python
plt.bar(labels, values, width=0.5)
plt.xticks(rotation=45)
plt.yticks(rotation=45)
```

```python
labels = ['A', 'B', 'C']
values = [190, 187, 184]
ticks= ['1', '2', '3']

plt.bar(labels, values)
plt.xticks(labels, ticks)
```
