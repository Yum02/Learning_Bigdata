9. 막대 그래프 심화

```python
import matplotlib.pyplot as plt
```

```python
labels = ['A', 'B', 'C']
values = [190, 187, 184]

plt.barh(labels, values)
plt.xlim(175, 195)
```

```python
bar = plt.bar(labels, values)
bar[0].set_hatch('/')
bar[1].set_hatch('x')
bar[2].set_hatch('..')
```

```python
bar = plt.bar(labels, values)
plt.ylim(175, 195)

for idx, rect in enumerate(bar):
    plt.text(idx, rect.get_height(), values[idx], ha='center', color='b')
```
