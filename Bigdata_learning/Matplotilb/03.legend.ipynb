```python
import matplotlib.pyplot as plt
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y, label="data")
plt.legend()
```

```python
plt.plot(x, y, label="data")
plt.legend(loc='upper right')
```

```python
plt.plot(x, y, label="data")
plt.legend(loc='lower right')
```

```python
plt.plot(x, y, label="legent")
plt.legend(loc=(0.5, 0.5))
```
