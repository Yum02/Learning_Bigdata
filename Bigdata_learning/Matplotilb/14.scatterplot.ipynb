14. 산점도 그래프

```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_excel('../Pandas/score1.xlsx')
df
```

```python
df['Grade'] = [3, 3, 2, 1, 1, 3, 2, 2]
df
```

```python
plt.scatter(df['EN'], df['Math'])
plt.xlabel('EN score')
plt.ylabel('Math score')
```

```python
import numpy as np
sizes = np.random.rand(8) * 1000
sizes
```

```python
plt.scatter(df['EN'], df['Math'], s = sizes)
plt.xlabel('EN score')
plt.ylabel('Math score')
```

```python
sizes = df['Grade'] * 100

plt.scatter(df['EN'], df['Math'], s = sizes)
plt.xlabel('EN score')
plt.ylabel('Math score')
```

```python
plt.scatter(df['EN'], df['Math'], s = sizes, c=df['Grade'], cmap='viridis', alpha=0.3)
plt.xlabel('EN score')
plt.ylabel('Math score')
```

```python
plt.figure(figsize=(10, 10))

plt.scatter(df['EN'], df['Math'], s = sizes, c=df['Grade'], cmap='viridis', alpha=0.5)
plt.xlabel('EN score')
plt.ylabel('Math score')
plt.colorbar(ticks=[1, 2, 3], label='Grade', shrink=0.3, orientation='horizontal')
```
