```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y)
plt.title('Line graph')
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
```

```python
plt.plot(x,y)
plt.xlabel('X', color='red')
plt.ylabel('Y', color='#00aa00')
```

```python
plt.plot(x,y)
plt.xlabel('X', color='red', loc='right')
plt.ylabel('Y', color='#00aa00', loc='top')
```

```python
plt.plot(x, y)
plt.xticks([1, 2, 3])
plt.yticks([3, 6, 9, 12])
plt.show()
```
